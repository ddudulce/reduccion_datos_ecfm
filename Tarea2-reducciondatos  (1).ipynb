### Cálculo de la media, desviación estandar, valor máximo y mínimo de 'coolant' y 'pm'

```python
temperature_cols = ['coolant', 'pm']
stats_summary = {}

for col in temperature_cols:
    mean_val = data[col].mean()
    std_val = data[col].std()
    max_val = data[col].max()
    min_val = data[col].min()

    print(f"\nColumna: {col}, Media: {mean_val:.2f}, Desviación Estándar: {std_val:.2f}, Máximo: {max_val:.2f}, Mínimo: {min_val:.2f}")
```

```python
profile_means = data.groupby("profile_id")[temperature_cols].mean()
```

```python
profiles = data["profile_id"].unique()
for profile in profiles:
    mean_coolant = data[data["profile_id"] == profile]["coolant"].mean()
    mean_pm = data[data["profile_id"] == profile]["pm"].mean()
    print(f"Perfil {profile}: Coolant - {mean_coolant:.2f}, PM - {mean_pm:.2f}")
```
